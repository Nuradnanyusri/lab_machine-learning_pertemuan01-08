### 1.1 Membuat Array dengan 20 Angka Random (0-100)

```python
# Set random seed untuk reproducibility
np.random.seed(42)

# Membuat array dengan 20 angka random antara 0-100
random_array = np.random.randint(0, 101, size=20)

print("Array dengan 20 angka random (0-100):")
print(random_array)
print(f"\nShape: {random_array.shape}")
print(f"Data type: {random_array.dtype}")
```

### 1.2 Menghitung Statistik Deskriptif

```python
# Menghitung berbagai statistik
mean_val = np.mean(random_array)
median_val = np.median(random_array)
std_val = np.std(random_array)
min_val = np.min(random_array)
max_val = np.max(random_array)

print("="*50)
print("STATISTIK DESKRIPTIF")
print("="*50)
print(f"Mean (Rata-rata)        : {mean_val:.2f}")
print(f"Median (Nilai Tengah)   : {median_val:.2f}")
print(f"Standard Deviation      : {std_val:.2f}")
print(f"Minimum                 : {min_val}")
print(f"Maximum                 : {max_val}")
print(f"Range                   : {max_val - min_val}")
print("="*50)
```

### 1.3 Membuat Matrix 5x5 dengan Angka Random

```python
# Membuat matrix 5x5 dengan angka random
matrix_5x5 = np.random.randint(1, 100, size=(5, 5))

print("Matrix 5x5 dengan angka random:")
print(matrix_5x5)
print(f"\nShape: {matrix_5x5.shape}")
print(f"Total elements: {matrix_5x5.size}")
```

### 1.4 Menghitung Transpose Matrix

```python
# Menghitung transpose
matrix_transpose = matrix_5x5.T

print("Matrix Original:")
print(matrix_5x5)
print("\nMatrix Transpose:")
print(matrix_transpose)

# Verifikasi transpose
print("\nVerifikasi: Element [0,1] original = Element [1,0] transpose?")
print(f"Original [0,1] = {matrix_5x5[0,1]}")
print(f"Transpose [1,0] = {matrix_transpose[1,0]}")
print(f"Match: {matrix_5x5[0,1] == matrix_transpose[1,0]}")
```

### 1.5 Menghitung Determinant Matrix

```python
# Menghitung determinant
determinant = np.linalg.det(matrix_5x5)

print("Matrix 5x5:")
print(matrix_5x5)
print(f"\nDeterminant: {determinant:.2f}")

# Interpretasi determinant
if abs(determinant) < 1e-10:
    print("\n📊 Interpretasi: Matrix singular (tidak invertible)")
else:
    print("\n📊 Interpretasi: Matrix non-singular (invertible)")
    print(f"   Matrix dapat di-inverse karena determinant ≠ 0")
```

## 📝 TUGAS 2: Analisis Data dengan Pandas (30 poin)

Membuat DataFrame mahasiswa dan melakukan analisis data

### 2.1 Membuat DataFrame Data Mahasiswa

```python
# Data mahasiswa (10 mahasiswa)
data_mahasiswa = {
    'NIM': ['105841115623', '105841115624', '105841115625', '105841115626', '105841115627', 
            '105841115628', '105841115629', '105841115630', '105841115631', '105841115632'],
    'Nama': ['Nur Adnan Yusri', 'Fatahillah Furqon', 'Nahcrief', 'Muh Akram Riyad', 
             'Suriani', 'Nurul Afsari', 'Andi Mawadda Taiba', 'Nurul Ilmi Aulia',
             'Wardiman Saputra J', 'Ferdiansyah Saputra'],
    'Nilai_Matematika': [85, 92, 78, 88, 95, 82, 90, 87, 91, 84],
    'Nilai_Fisika': [80, 88, 75, 85, 90, 78, 92, 83, 89, 81],
    'Nilai_Kimia': [88, 90, 80, 92, 93, 85, 89, 91, 87, 86]
}

df_mahasiswa = pd.DataFrame(data_mahasiswa)

print("="*70)
print("DATA MAHASISWA")
print("="*70)
print(df_mahasiswa)
print("\n" + "="*70)
```

### 2.2 Menghitung Rata-rata Nilai Per Mata Kuliah

```python
# Rata-rata per mata kuliah
avg_matematika = df_mahasiswa['Nilai_Matematika'].mean()
avg_fisika = df_mahasiswa['Nilai_Fisika'].mean()
avg_kimia = df_mahasiswa['Nilai_Kimia'].mean()

print("="*50)
print("RATA-RATA NILAI PER MATA KULIAH")
print("="*50)
print(f"Matematika : {avg_matematika:.2f}")
print(f"Fisika     : {avg_fisika:.2f}")
print(f"Kimia      : {avg_kimia:.2f}")
print("="*50)

# Membuat DataFrame ringkasan
df_avg_matkul = pd.DataFrame({
    'Mata Kuliah': ['Matematika', 'Fisika', 'Kimia'],
    'Rata-rata': [avg_matematika, avg_fisika, avg_kimia]
})

print("\nRingkasan dalam bentuk tabel:")
print(df_avg_matkul)
```

### 2.3 Menghitung Rata-rata Nilai Per Mahasiswa

```python
# Menghitung rata-rata per mahasiswa
df_mahasiswa['Rata_rata'] = df_mahasiswa[['Nilai_Matematika', 'Nilai_Fisika', 'Nilai_Kimia']].mean(axis=1)

print("DATA MAHASISWA DENGAN RATA-RATA NILAI:")
print(df_mahasiswa[['NIM', 'Nama', 'Rata_rata']].to_string(index=False))

# Statistik rata-rata
print(f"\nRata-rata tertinggi: {df_mahasiswa['Rata_rata'].max():.2f}")
print(f"Rata-rata terendah: {df_mahasiswa['Rata_rata'].min():.2f}")
print(f"Rata-rata keseluruhan: {df_mahasiswa['Rata_rata'].mean():.2f}")
```

### 2.4 Menampilkan 5 Mahasiswa dengan Nilai Tertinggi

```python
# Urutkan berdasarkan rata-rata (descending) dan ambil 5 teratas
top_5_mahasiswa = df_mahasiswa.nlargest(5, 'Rata_rata')[['NIM', 'Nama', 'Rata_rata']]

print("="*60)
print("🏆 TOP 5 MAHASISWA DENGAN NILAI TERTINGGI")
print("="*60)
for idx, row in enumerate(top_5_mahasiswa.itertuples(index=False), 1):
    print(f"{idx}. {row.Nama:<20} (NIM: {row.NIM}) - Rata-rata: {row.Rata_rata:.2f}")
print("="*60)
```

### 2.5 Filter Mahasiswa dengan Rata-rata ≥ 80

```python
# Filter mahasiswa dengan rata-rata >= 80
mahasiswa_lulus = df_mahasiswa[df_mahasiswa['Rata_rata'] >= 80]

print("="*70)
print("MAHASISWA DENGAN RATA-RATA NILAI ≥ 80")
print("="*70)
print(mahasiswa_lulus[['NIM', 'Nama', 'Rata_rata']].to_string(index=False))
print(f"\nTotal mahasiswa: {len(mahasiswa_lulus)} dari {len(df_mahasiswa)} mahasiswa")
print(f"Persentase: {(len(mahasiswa_lulus)/len(df_mahasiswa)*100):.1f}%")
```

### 2.6 Export DataFrame ke CSV

```python
# Export ke CSV
filename = 'data_mahasiswa_pertemuan01.csv'
df_mahasiswa.to_csv(filename, index=False)

print(f"✓ Data berhasil di-export ke file: {filename}")

# Verifikasi dengan membaca kembali
df_verify = pd.read_csv(filename)
print(f"\n✓ Verifikasi: File berisi {len(df_verify)} baris data")
print("\nPreview 3 baris pertama dari file CSV:")
print(df_verify.head(3))
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df_mahasiswa['Nilai_Matematika'], df_mahasiswa['Nilai_Fisika'],
           s=150, c=df_mahasiswa['Rata_rata'], cmap='viridis', 
           edgecolor='black', linewidth=1.5, alpha=0.7)

# Tambahkan colorbar
cbar = plt.colorbar()
cbar.set_label('Rata-rata Nilai', rotation=270, labelpad=20, fontweight='bold')

# Tambahkan trend line
z = np.polyfit(df_mahasiswa['Nilai_Matematika'], df_mahasiswa['Nilai_Fisika'], 1)
p = np.poly1d(z)
plt.plot(df_mahasiswa['Nilai_Matematika'], p(df_mahasiswa['Nilai_Matematika']),
         "r--", linewidth=2, label='Trend Line')

# Tambahkan label untuk beberapa titik
for idx, row in df_mahasiswa.iterrows():
    if row['Rata_rata'] >= 90:  # Label mahasiswa dengan rata-rata tinggi
        plt.annotate(row['Nama'].split()[0], 
                    (row['Nilai_Matematika'], row['Nilai_Fisika']),
                    xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.title('Hubungan Nilai Matematika dan Fisika', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Nilai Matematika', fontsize=12, fontweight='bold')
plt.ylabel('Nilai Fisika', fontsize=12, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# Hitung korelasi
correlation = df_mahasiswa['Nilai_Matematika'].corr(df_mahasiswa['Nilai_Fisika'])
print("📊 Interpretasi:")
print(f"   - Koefisien korelasi: {correlation:.3f}")
if correlation > 0.7:
    print("   - Terdapat korelasi positif yang kuat antara Matematika dan Fisika")
elif correlation > 0.4:
    print("   - Terdapat korelasi positif yang moderat antara Matematika dan Fisika")
else:
    print("   - Korelasi antara Matematika dan Fisika relatif lemah")
```
